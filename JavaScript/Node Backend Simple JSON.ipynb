```javascript
/*when you type in the url, you are making a request to server,
the server hears that request, and it knows what to do because somebody
coded it and it knows how to respond
you, as the developer, gets to decide what happens.
the server listens for specific things. we get to decide what happens
200 page means everything loaded okay

*/


//Modules
const http = require('http');
const fs = require('fs')
const url = require('url');
const querystring = require('querystring');
const figlet = require('figlet')                //ascii 404 page






const server = http.createServer((req, res) => {
  const page = url.parse(req.url).pathname;
  const params = querystring.parse(url.parse(req.url).query);
  console.log(page);
  if (page == '/') {                                    //server sends a request on the main page
    fs.readFile('index.html', function(err, data) {         //if somebody goes to the main page, send them this file
      res.writeHead(200, {'Content-Type': 'text/html'});
      res.write(data);
      res.end();
    });
  }
  else if (page == '/otherpage') {                     //server sends a request on /otherpage
    fs.readFile('otherpage.html', function(err, data) {   //if somebody goes to this page, send them this file
      res.writeHead(200, {'Content-Type': 'text/html'});
      res.write(data);
      res.end();
    });
  }
  else if (page == '/otherotherpage') {                      //server sends a request on /otherotherpage
    fs.readFile('otherotherpage.html', function(err, data) {    //if somebody goes to this page, send them this file
      res.writeHead(200, {'Content-Type': 'text/html'});
      res.write(data);
      res.end();
    });
  }
  else if (page == '/api') {                              //server listens for pages with /api in it
    if('student' in params){                               
      if(params['student']== 'leon'){
        res.writeHead(200, {'Content-Type': 'application/json'});
        const objToJson = {
          name: "leon",
          status: "Boss Man",
          currentOccupation: "Baller"
        }
        res.end(JSON.stringify(objToJson));
      }//student = leon
      else if(params['student'] != 'leon'){
        res.writeHead(200, {'Content-Type': 'application/json'});
        const objToJson = {
          name: "unknown",
          status: "unknown",
          currentOccupation: "unknown"
        }
        res.end(JSON.stringify(objToJson));
      }//student != leon
    }//student if
  }//else if
  else if (page == '/css/style.css'){                              //server listens for a url with a stylesheet
    fs.readFile('css/style.css', function(err, data) {
      res.write(data);
      res.end();
    });
  }else if (page == '/js/main.js'){                                     //server listens for a url with js file
    fs.readFile('js/main.js', function(err, data) {
      res.writeHead(200, {'Content-Type': 'text/javascript'});
      res.write(data);
      res.end();
    });
  }else{
    figlet('404!!', function(err, data) {                  //if server doesnt have the url listed above, it responds with a 404
      if (err) {
          console.log('Something went wrong...');
          console.dir(err);
          return;
      }
      res.write(data);
      res.end();
    });
  }
});

server.listen(8000);               // go to localhost:8000
```
