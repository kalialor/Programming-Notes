<h3>What is Node.js?</h3>
<br>

Node is a JavaScript Runtime. (Not a language or a framework). Built on the V8 JavaScript engine. (Same as Google Chrome). 
Written in C++ 
Essentially allows us to run JavaScript code on the server.
<br><br>


<h3>Why use Node?</h3>

Fast, efficient and highly scable. Event driven, non-blocking I/O model. Popular in the industry. Same language on the front and back end (JS).
<br><br>

<h3>Non-blocking I/O</h3>

Works on a single thread using non-blocking I/O (input/output) calls
Supports tens of thousands concurrent connections. Optimizes throughput & scalability in apps with many I/O operations. All of this makes Node.js apps very fast & efficient. 
<br><br>

<h3>Node's Event Loop</h3>

Single threaded 
Supports concurrency via events & callbacks
EventEmitter class is used to bind events and listeners
<br><br>

<h3>Best Types of Projects for Node:</h3>

REST API & Microservices
Real Time Services (Chat, Lives Updates)
CRUD apps-blogs, shopping carts, social networks
Tools & Utilities

Short Answer: Anything that is not CPU intensive
<br><br>

<h3>URL Module</h3><br>
<strong>Instructions:</strong>create a file called url_demo.js<br>
type in terminal `node url_demo` for each command

```javascript
const url = require('url');

const myURL = new URL('http://mywebsite.com/hello.html?id =100&status=active');



//Serialized URL
console.log(myUrl.href);
console.log(myUrl.toString());


//Host (root domain)
console.log(myUrl.host);


//Hostname (does not get port)
console.log(myUrl.hostname);



//Pathname
console.log(myUrl.pathname);


//Serialized Query
console.log(myUrl.searh);


//Params Object      <---creates an object
console.log(myUrl.searchParams);




//Add param
myUrl.searchParams.append('abc', '123');
console.log(myUrl.searchParams);

//Result: is the params object, plus the abc, 123



//Loop through params
myUrl.searchParams.forEach(value, name) => console.log(`${name}: ${value}`));
```

above is not very efficient<br>
here's a better way:
<br><br><br><br>

```javascript
const http = require('http');            //creating a server, that'll get a request & response
const path = require('path');
const fs = require('fs');

const server = http.createServer((req,res) => { 

//Build file path
let filePath = path.join(
    -dirname,                                 //directory name
    'public',                                 //look in the public folder
    req.url === '/' ? 'index.html' : req.url  //request.url. if it's / we are going to load the index
);                                            //if not, we are going to load whatever req.url is called
    
    //Extension of file
    let extname = path.extname(filePath);
    
    //Initial content type
    let contentType = 'text/html';
    
    //Check ext and set content type
    switch (extname) {
        case '.js':
            contentType = 'text/javascript';
            break;
        case 'css':
            contentType = 'text/css';
            break;
        case '.json':
            contentType = 'application/json';
            break;
        case '.png':
            contentType = 'image/png';
            break;
        case '.jpg':
            contentType = image/jpg;
            break;
    }

//Read File
fs.readFile(filePath, (err,content) =>{
    if(err) {
        if(err.code == 'ENDENT') {        //means not found
            //PAGE NOT FOUND
        fs.readFile(path.join(_dirname, 'public', '404.html'),
            err, content) => {
            res.writeHead(200, {'Content-Type': 'text/html'});
            res.end(content, 'utf8');
        }) 
        }else {
            //Some server error
        res.writeHead(500);
        res.end(`Server Error: ${err.code}`);
        }
    } else {
        //Success
        res.writeHead(200, {'Content-Type': contentType});
        res.end(content, 'utf8');         //sends the content of the file
    }
});

const PORT = process.env.PORT || 5000;

server.listen(PORT, () => console.log(`Server running on port ${PORT}`));
```
