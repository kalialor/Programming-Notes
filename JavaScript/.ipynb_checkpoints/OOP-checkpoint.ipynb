<h1>OBJECT</h1>

an object is a collection of variables and functions 

variables are called properties 
functions are called methods 



<h3>EXAMPLE:</h3>

```javascript
function Dog(dogSize, dogBreed, dogHair, dogEyeColor){
    this.dogSize = dogSize               //the value on the right side is the parameter
    this.dogBreed = dogBreed            //the value on the left is the property
    this.dogHair = dogHair              
    this.dogEyeColor = dogEyeColor
    this.bark = function(){
        console.log('WOOOOF!')
    }
    this.run = function(){
        console.log('THIS IS SO MUCH FUN!!!!')
    }
    this.sleep = function(){
        console.log('zzzzzzZZZZZZZzzzzz')
    }
}

let Oliver = new Dog('medium', 'pomsky', 'long haired', 'blue')
```

<strong>Encapsulation</strong> is the process of storing functions (methods) with their associated data
(properties) in one thing (object)


What is the reason we use encapsulation?
    makes code more readable
    orgainizes code
    makes it easier to add new stuff
    makes it so you were not afraid to make changes

<strong>ABSTRACTION</strong> is all about hiding the details and showing only the essentials
            simple, predictable, managable




<br><br>
Why do we use it?
    smaller more manageable pieces of code 
    helps you split the complexity your software project into manageable parts

<strong>INHERITANCE</strong> 

    inheritating from parent object
    we just eliminated a bunch of redundant code
    make a class from another class for a hierarchy of classes that share a set of 
    properties and methods
    create new objects based on old ones


    Why do we use it?
        keeps code organized, reduces duplication 


    the keyword super allows us to grab all the stuff that is in the parent object and use it for 
    the child object                             see class 31
    can extend and grab what it needs
    we are not redoing the parts we don't need




<br><br><br>



<strong>POLYMORPHISM</strong>

    a piece of code that automagically works with a bunch of
    different objects
    
    provides a way to perform a single action in different forms. 
    provides an ability to call the same method on different JavaScript objects

    helps you avoid if/else and switch cases

<br><br>

```javascript
//underscore tells to other developers not to change the name property
//it doesn't do anything. it just a sign.
//DO NOT TOUCH UNDERSCORE
//you can't access underscore outside of the class

//anything that comes out of the dog class gets the benefit of the animal class

//everytime you have a underscore you need a getter

//whenever you want to use something from the parent class, you the word super




class Animal{
    constructor(name){
        this._name = name
    }
    get name(){
        return this._name
    }
    speak(){
        console.log(`${this._name} makes a sound`)
    }
}
class Dog extends Animal{
    constructor(name,breed){
        super(name)
        this._breed = breed
    }
    get breed(){
        return this._breed
    }
    speak(){
        super.speak()
        console.log(`${this.name} barks`)
    }    
}
class Cat extends Animal{
    constructor(name,breed){
        super(name)
        this._breed = breed
    }
    get breed(){
        return this._breed
    }
    speak(){
        super.speak()
        console.log(`${this.name} meows`)
    }    
}

let simba = new Dog('Simba','Shepard')
let machi = new Dog('The Machine','Pitbull')
let salem = new Cat('Salem', 'American Shorthair')

let farm = [simba,machi,salem]

for( a of farm ){
    a.speak()
}
```

ANOTHER EXAMPLE:

```javascript
// Come up with with a parent class
// Extend that parent class into two children
// Use Encapsulation, Abstraction, Inheritance, and Polymorphism 
// This is encapsulation because we infused properties and methods in one spot


class Contractor{                        //object
    constructor(name,role){              //properties
        this._name = name                //abstraction, the underscore
        this._role = role
    }
    get name(){                  //you need the get keyword for abstraction
        return this._name
    }
    get role(){
        return this._role
    }
    sayHello(){                      //method
        console.log(`Hello, my name is ${this._name} on the ${this._role} team at #100Devs!`)
    }
}
class Front extends Contractor{      //child object
    constructor(name,role,tech){
        super(name,role)             //inheritance, go up a level, gives us access to the parent class 
        this._tech = tech           //notice how we didn't have to write name and role again
    }
    get tech(){
        return this._tech
    }
    sayHello(){
        console.log(`Hello, I am on the ${this._role} team at #100Devs and I use ${this._tech}`)
    }
}
class Back extends Contractor{
    constructor(name,role,tech){
        super(name,role)
        this._tech = tech
    }
    get tech(){
        return this._tech
    }
    sayHello(){
        console.log(`Hello, I am on the ${this._role} team at #100Devs and I use ${this._tech}`)
    }
}
let kalia = new Contractor('Kalia','Front-end')
let simba = new Front('Simba','Front-end','React')
let machi = new Back('The Machine','Back-end','Node')

let agency = [kalia,simba,machi]           //polymorphism

for(person of agency){          //this is going to loop through the agency list 3 times           
    person.sayHello()           //person can be anything. it is a variable name
}                               //so each person will sayHello 
```
