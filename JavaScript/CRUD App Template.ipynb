```javascript
//TEMPLATE


//Declare variables
const express = require('express')
const app = express()
const cors = require('cors')
const PORT = 8500;
const mongoose = require('mongoose')
//const MongoClient = require('mongodb').MongoClient    <---if you're using Mongo instead of Mongoose
require('dotenv').config()


//Set Middleware
app.set("view engine", "ejs")       //sets up ejs
app.use(express.static('public'))  //tells express if you need stylesheets then look in the public folder
app.use(express.urlencoded({extended:true})) //our url parser. it helps validate the information being passed back and forth. extended property helps us pass arrays to the server



mongoose.connect(process.env.DB_CONNECTION,
    {useNewUrlParser: true},
    () => {console.log('Connected to db!')}
)


/*IF USING MONGODB

let db, 
    dbConnectionString = process.env.DB_STRING, 
    dbName = 'NAME OF DATABASE FIND ON MONGOATLAS',
    collection

MongoClient.connect(dbConnectionString)
    .then(client => {
        console.log('Connected to Database')
        db  client.db(dbName)
        collection = db.collection('TYPE IN COLLECTION NAME HERE')
    })
    
*/


app.listen(PORT, () => console.log(`Server is running on port ${PORT}`))
```
