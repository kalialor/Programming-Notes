      <strong>Box Model</strong>           


<br><br>
Everything is a box

<br><br>
<strong>padding</strong>-increases the distance between the element and the border. padding pushes the border away from the element<br><br><br><br>

<strong>margin</strong>-pushes the box model around. 
<br><br><br><br>


<strong>Floats</strong>

Elements have a tendancy to move up and into a corner <br>
<br><br>


<br><br>


-box-sizing ignores border when calculating width because with border then the size of the section elements will be over 100%

-clear: both tells header & footer to come after it

-float: left causes the sections to sit next to each other


*{
    box-sizing: border-box;
}

header, footer{
    height:100px;
    border: 3px solid black;
    clear: both;                   
}
section{
    height: 300px;
    width: 33.3333%;
    border 3px solid black;
    float: left;
}

<br><br><br><br>
<strong>Font size: em & rem</strong>
<br><br>

<strong>em</strong>-get their font size from the parent element. You want to use this when you have a lot of child elements. Instead of changing each pixel you only need to change the parent element's pixel and the child elements will follow suit<br><br>

<strong>rem</strong>-stands for root em. They only get their font size from the HTML document. used for accessiblity. 
<br><br><br>
Example:


html{
    font-size: 62.5%;
    }
section{
    font-size: 10px;
}
p{
    font-size: 1em;
}
vs.
p{
    font-size: 1rem;
}

<strong>Flexbox</strong><br>
<br>


display flex: causes the elements to sit next to each other

justify-content: space-around. gives space around each element

flex-basis: is used to specify the initial size of the flex element. 
        gives more space to the individual elements
        values: number, auto, initial, inherit
      
flex-wrap: wrap; have the elements go to the next line
        



nav,ul{
    display: flex;
    justify-content: space-around;
}

ul{
    flex-basis: 30%;
}
section {
    flex-wrap: wrap;
