<h2>What is Git?</h2>
<br>

    git is a verision control system. a fancy way of saying it tracks your changes.




https://docs.github.com/en/get-started/writing-on-github/getting-started-with-writing-and-formatting-on-github/basic-writing-and-formatting-syntax

 
 <br><br>

<strong>Repositories</strong>-a repo is a collection of commits, each of which is an archive of what the project's
             working tree looked like at a past date, whether on your machine or someone else's. 
             a container for a project that is tracked by git. 

<br><br>
    2 types of repos:

                local-an isolated repo stored on your own computer, where you can work on the local 
                verison of your project

                remote repo-generally stored outside of your isolated local system, usually on a remote
                     server. it is especially useful when working in teams. 
                     this is the place where you can share your project's code, see other people's
                    code and integrate it into your local version of the project, also push 
                    your changes to the remote repo.



 <br><br>

<strong>Git init</strong>-initializes changes of a file or project

         use    ls -a     and the .git file should show up


        to create a new repo and start tracking your project with git. this folder
        is where git stores all internal tracking data for the current repo.



 <br><br>


<strong>Git add</strong>-adds files ready to be committed. from the project folder, we can use git add command 
        to add our files to the staging area, which allows them to be tracked. 

        example: git add file.js

        
        if you want to add all the files inside of the project folder to the staging area:
                git add . 




 <br><br>


<strong>Git status</strong>-view commits



 <br><br>

<strong>commit</strong>-a checkpoint. committing is the process in which the changes are offically added to the git
       repo. 
       in git, we can consider commits to be checkpoints, or snapshots of your project at its current
       state. In other words, we basically save the current version of our code in a commit. We can 
       create as many commits as we need in the commit history, and we can go back and forth between 
       commits to see the different revisions of our project's code. That allows use to efficiently 
       manage our progress and track the project as it gets developed. 

       commits are usually created at logical points as we develop our project, usually after adding in
       the specific contents, features, or modifications (like new functionalities or bug fixes, for 
       example)

       ****before we can commit our code, we need to place it in the staging area. 


How to commit:

                git commit -m "type updates here"           <----- m stands for message
                                                                   

                                                                
 <br><br>

<strong>Git push command:</strong> 


the git push command takes two arguments: 
    -a remote name, for example origin
    -a branch name, for example main

       ex:     git push remote-name branch-name

                git push origin main      common for pushing from local to remote repo

 <br><br>


<strong>git log</strong>-to see all the commits that we made for our project. The logs will show details for each commit. author, hash,
        date & time, & commit message.
                                                                   
                                                                   
  
                                                                   
                                                                   
                                                                   
    <br><br>                                                                
                                                                   
                                                                   
<strong>git checkout</strong>-replace the <commit hash> with the actual hash for the specific commit that you want to visit, which is listed
in the "git log" command





 <strong>git checkout master</strong>-to go to the latest comment (the newest version of our project code), you can type this command:
                                        git checkout master
     



 <br><br>



    <strong>.gitignore</strong>-to ignore files that you don't want to be tracked or added to the staging area, you can create a file called 
           .gitignore in your main project folder. Inside of that file, you can list all the files and folder names that you
            definitely do not want to track (each ignored file and folder should go to a new line inside the .gitignore file


 <br><br>
 <br><br>

 <br><br>






    <strong>Branch</strong>-A branch can be interpreted as an individual timeline of our project's commits. With Git, we can create many of these
       alternative environments (i.e. we can create different branches) so other versions of our project code can exist and
can be tracked in parallel. That allows us to add new (experimental, unfinished, and potentially buggy) features in separate branches, without touching the 'official' stable version of our project code. (Which is usually kept on the master branch). When we initialize a repo and start making commits, they are saved to the MASTER branch by default. 

 <br><br>



        Creating a new Branch

            ``git branch <new-branch-name>``
     <br><br>
               
            The new branch that gets created will be the reference to the current state of you repo. 
                
            It is a good idea to create a development branch where you an work on improving your code, adding new experimental 
            features, and similar. After development and testing these new features to make sure they don't have any bugs and 
            they can be used, you can merge them to the master branch. 

 <br><br>

    Renaming branches

        To rename a branch, you'd use the same git push command, but you would add one more 
        argument: the name of the new branch. For example: 
            git push remote-name local-branch-name: remote-branch-name

            This pushes the local-branch-name to your remote-name, but it is renamed to 
            remote-branch-name

    <h2>Git Clone vs Git Fork</h2>
<br><br>
https://www.youtube.com/watch?v=D2rbsqGCi2o<br><br>
Depends on how much control a developer has over a repository.
<br><br><br>
    <h3>Git Clone</h3><br>
git clone feature has no rights to the repo. you can't push. It won't let you unless the person who made the repo gives you permission.If you are working in a team, this would be a good option because you know the owner.<br><br>
How to clone a repo:<br>
    git clone url-of-the-repo
<br><br><br>
    <h3>Git Fork:</h3><br><br>
If you don't know the owner of the repo, then use this option. That way you can have the copy of the repo and make changes to your own copy.
    Click on fork button on GitHub, then clone the repo, so it shows up on your local machine. Make changes, then you can push changes. It will push changes to your forked copy.
